```python
def get_distance(t1,n1,t2,n2):
    df1 = df_loc[(df_loc["Type"]==t1)&(df_loc["ID"]==n1)]
    df2 = df_loc[(df_loc["Type"]==t2)&(df_loc["ID"]==n2)]
    distance = np.abs(df1.iloc[0].x - df2.iloc[0].x) + np.abs(df1.iloc[0].y - df2.iloc[0].y)
    return distance
```

```python
import itertools

def findsubsets(s, n):
    return list(itertools.permutations(s, n))

def get_min_distance(route):
    subs = findsubsets(route, len(route))
    mindist = 99999
    opt = []
    wo = -1
    for w in range(1,3):
        for r in subs:
            dist = get_distance("Shop",r[0],"Warehouse",w) + get_distance("Shop",r[len(r)-1],"Warehouse",w)
            for i in range(len(r)-1):
                dist += get_distance("Shop",r[i],"Shop",r[i+1])
            if dist<mindist:
                mindist = dist
                opt = r
                wo = w
    return (mindist, opt, wo)
```

```python
# @jit
def flatten(list_of_lists):
    if len(list_of_lists) == 0:
        return list_of_lists
    if isinstance(list_of_lists[0], list):
        return flatten(list_of_lists[0]) + flatten(list_of_lists[1:])
    return list_of_lists[:1] + flatten(list_of_lists[1:])
```

```python
from IPython.display import clear_output

mindist = 9999
optdist = []
optc = {}
items = 0
count = 0
limit = 1000000
def findcombs(s):
    global mindist
    global optdist
    global optc
    global items
    global count
    global limit
    
    comb = list(itertools.combinations(range(len(s)), 2))
    for i in comb:
        if items>limit:
            return 0
        newc = []
        n = s[i[0]]+s[i[1]]
        if tuple(n) in opt_route.keys():                
            newc.append(n)
            for j in s:
                if j[0] not in flatten(n):
                    newc.append(j)
            newc.sort()
            if tuple(map(tuple, newc)) not in optc.keys():
                dist = 0
                for n in newc:
                    dist += opt_route[tuple(n)][0]
#                 print(newc, dist, end="")
                if (dist < mindist) | ((dist == mindist) & (len(newc) < len(optdist))):
                    mindist = dist
                    optdist = newc
#                 else:
#                     print()
    #                 count += 1
    #                 if count >= 100000:
    #                     print(".",end="")
    #                     count = 0
                print(optdist, mindist, end=" << MIN\n")
                print(newc, dist, items)
                clear_output(wait=True)
        
                items += 1
                optc[tuple(map(tuple, newc))] = dist
                findcombs(newc)

cc = []
for i in range(1,maxs+1):
    cc.append([i])

start = time.time()
findcombs(cc)
print("Elapsed time:",time.time()-start)
print("Total combinations:",items)
```

```python
n = 20
rnd = np.random
rnd.seed(0)
```

```python
N = [i for i in range(0, n)]
V = N + [20] + [21]
A = [(i, j) for i in V for j in V if i != j]
c = {(i, j): df_dist[i][j] for i, j in A}
Q = 20
q = {i: round(df_days_months[["January"]].T["S"+str(i+1)]["Total"][0],1) for i in N}
```

```python
c
```

```python
q
```

```python
from gurobipy import Model, GRB, quicksum
```

```python
mdl = Model('CVRP')
```

```python
x = mdl.addVars(A, vtype=GRB.BINARY)
u = mdl.addVars(N, vtype=GRB.CONTINUOUS)
```

```python
A
```

```python
mdl.modelSense = GRB.MINIMIZE
mdl.setObjective(quicksum(x[i, j]*c[i, j] for i, j in A))
```

```python
mdl.addConstrs(quicksum(x[i, j] for j in V if j != i) == 1 for i in N)
mdl.addConstrs(quicksum(x[i, j] for i in V if i != j) == 1 for j in N)
mdl.addConstrs((x[i, j] == 1) >> (u[i]+q[j] == u[j])
               for i, j in A if i not in [20,21] and j not in [20,21])
mdl.addConstrs(u[i] >= q[i] for i in N)
mdl.addConstrs(u[i] <= Q for i in N)
```

```python
mdl.Params.MIPGap = 0.1
mdl.Params.TimeLimit = 999  # seconds
mdl.optimize()
```
